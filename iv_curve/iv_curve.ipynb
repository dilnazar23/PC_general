```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def PlotPV(current_file,voltage_file,folder):
    cur = np.loadtxt(current_file,skiprows=1, delimiter=",",usecols=(1,2))     
    cur_in = cur[:,0]
    cur_out = cur[:,1]
        
    vol = np.loadtxt(voltage_file,skiprows=1, delimiter=",",usecols=(1,2))    
    vol_in = vol[:,0]
    vol_out = vol[:,1]

    power_in = vol_in*cur_in*0.0001
    plt.figure(figsize=(12,4))
    plt.scatter(vol_in,cur_in,s=0.5,label=f"IV curve{folder}")
    plt.scatter(vol_in,power_in,s=0.5,label=f"PV curve{folder}")
    plt.grid()
    plt.axis([min(vol_in),max(vol_in),min(cur_in),max(cur_in)])
    plt.xlabel("Input Voltage(mV)")
    plt.ylabel("Input Current(mA)")
    plt.legend()
    plt.title("IV/PV curve scatter")
    #plt.show()
    
```

```python
plt.figure(figsize=(12,4))
for folder in os.listdir("Latest"):
    current_file=""
    voltage_file=""
    for file_name in os.listdir(os.path.join("Latest",folder)):
        if file_name.endswith("1a.csv"):
            current_file = os.path.join("Latest",folder,file_name)
        if file_name.endswith("1b.csv"):
            voltage_file = os.path.join("Latest",folder,file_name)
    cur = np.loadtxt(current_file,skiprows=1, delimiter=",",usecols=(1,2))     
    cur_in = cur[:,0]
    cur_out = cur[:,1]
        
    vol = np.loadtxt(voltage_file,skiprows=1, delimiter=",",usecols=(1,2))    
    vol_in = vol[:,0]
    vol_out = vol[:,1]

    power_in = vol_in*cur_in*0.000001

    # plt.scatter(vol_in,cur_in,s=0.5,label=f"IV curve{folder}")
    plt.scatter(vol_in,power_in,s=0.5,label=f"PV curve{folder}")
plt.grid()
#plt.axis([min(vol_in),max(vol_in),min(cur_in),max(cur_in)])
plt.xlabel("Input Voltage(mV)")
plt.ylabel("Input power(w)")
plt.legend()
plt.title("PV curve scatter")
    # plt.savefig(os.path.join(folder,".png"))
```
