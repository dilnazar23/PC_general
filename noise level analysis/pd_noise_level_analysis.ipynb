```python
import numpy as np
import matplotlib.pyplot as plt
import os
from scipy.stats import norm
from scipy.fft import fft, fftfreq
```

# Noise level characterisation

## Indoor noise

### Dark noise

```python
## Load data and plot gausian PDF distribution
indoor_dark_path = "indoor data\\dark"
for filename in os.listdir(indoor_dark_path):
    indoor_dark_data = np.loadtxt(os.path.join(indoor_dark_path,filename),delimiter=',',skiprows=1,usecols=(2,3,4,5))
    mean = np.mean(indoor_dark_data)
    std_dev = np.std(indoor_dark_data)
    # Generate x values
    x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)

    # Compute Gaussian PDF
    pdf = norm.pdf(x, mean, std_dev)

    # Plot histogram of the data
    #plt.hist(indoor_dark_data, bins=30, density=True, alpha=0.6, color='gray', label='Data Histogram')

    # Plot the Gaussian distribution curve
    plt.figure(figsize=(25
    ,5))
    plt.subplot(1,3,2)
    plt.plot(x, pdf, 'r-', label=f'Gaussian Fit\nMean={mean:.2f}, Std Dev={std_dev:.2f}')
    plt.title(f'Gaussian PDF of indoor dark data: {filename}')
    plt.xlabel('Data Values')
    plt.ylabel('Density')
    plt.legend()
    plt.grid()
    
        # Plot the raw data
    plt.subplot(1,3,1)
    plt.plot(indoor_dark_data)
    plt.title(f"raw data plot: {filename}")
    
    plt.grid()
    # Plot the frequency domain
    ## FFT Calculation
    fs = 50000
    n = len(indoor_dark_data)  # Number of samples
    fft_result = fft(indoor_dark_data)  # Perform FFT
    frequencies = fftfreq(n, 1/fs)  # Frequency bins
    magnitude = np.abs(fft_result) / n  # Normalize magnitude
    plt.subplot(1, 3, 3)
    plt.plot(frequencies[:n // 2], magnitude[:n // 2], label='each PD FFT')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Magnitude')
    plt.title('Frequency Domain plotting')
    plt.show()
```
